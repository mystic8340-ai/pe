```python
target=df[['Close']]
```

```python
Scaler=MinMaxScaler()
sca=Scaler.fit_transform(target)
```

```python
def create_dataset(dataset,time_step=60):
    X=[]
    y=[]
    for i in range(len(dataset)-time_step):
        X.append(dataset[i:(i+time_step),0])
        y.append(dataset[i+time_step,0])
    return np.array(X),np.array(y)
```

```python
X,y=create_dataset(sca,60)
```

```python
train_size=int(len(X)*0.8)
X_train=X[:train_size]
X_test=X[train_size:]
y_train=y[:train_size]
y_test=y[train_size:]
```

```python
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1],1)
X_test=X_test.reshape(X_test.shape[0],X_test.shape[1],1)
```

```python

rnn_model=Sequential()
rnn_model.add(SimpleRNN(50,
                        return_sequences=False,
                        input_shape=(60,1)))
rnn_model.add(Dropout(0.1))
rnn_model.add(Dense(1))
rnn_model.compile(
    optimizer='adam',    #(learnig_rate=learning_rate)
    loss='mse'
)
```

```python
rnn_model.summary()
```

```python
early_stops=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
history_rnn=rnn_model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=(X_test,y_test),
    callbacks=[early_stops]
)
    
```

```python
rnn_pred=rnn_model.predict(X_test)
```

```python
rnn_pred=Scaler.inverse_transform(rnn_pred.reshape(-1,1))
actual=Scaler.inverse_transform(y_test.reshape(-1,1))
```

```python
mse_rnn=mean_squared_error(actual,rnn_pred)
mae_rnn=mean_absolute_error(actual,rnn_pred)
print(mse_rnn)
print(mae_rnn)
```

```python
lstm_model=Sequential()
lstm_model.add(LSTM(units=50,
                        return_sequences=False))
lstm_model.add(Dropout(0.2))
lstm_model.add(Dense(1))
lstm_model.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
history_lstm=lstm_model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=(X_test,y_test),
    callbacks=[early_stops]
)
```

```python
lstm_pred=lstm_model.predict(X_test)
lstm_pred=Scaler.inverse_transform(lstm_pred.reshape(-1,1))
```

```python
mse_lstm=mean_squared_error(actual,lstm_pred)
mae_lstm=mean_absolute_error(actual,lstm_pred)
print(mse_lstm)
print(mae_lstm)
```

```python
comparison=pd.DataFrame({
    "Model":["SimpleRNN","LSTM"],
    "MSE":[mse_rnn,mse_lstm],
    "MAE":[mae_rnn,mae_lstm]})
comparison
```
